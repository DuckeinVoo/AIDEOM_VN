```python
import os
from pulp import *

# 1. Khởi tạo danh sách dự án P1 -> P15
P = list(range(1, 16))

# 2. Dữ liệu từ bảng 5.2
C = {1:12000, 2:11500, 3:18000, 4:4500, 5:3200, 6:5800, 7:6500, 8:15000, 
     9:2500, 10:7200, 11:4800, 12:8500, 13:20000, 14:3800, 15:1500}

C1 = {1:8500, 2:7500, 3:12000, 4:3500, 5:2500, 6:4000, 7:4500, 8:9000, 
      9:1800, 10:5000, 11:3500, 12:5500, 13:13000, 14:2800, 15:1200}

B = {1:21500, 2:20800, 3:32500, 4:9200, 5:6800, 6:11400, 7:12200, 8:28500, 
     9:5800, 10:13800, 11:8500, 12:16200, 13:35000, 14:7500, 15:3800}

# Tham số cộng hưởng giả định cho câu 5.5.c
B_extra = 5000 

# 3. Thiết lập mô hình MIP
m = LpProblem('VN_Synergy_Optimization', LpMaximize)

# Biến quyết định y[i] là biến nhị phân {0, 1}
y = LpVariable.dicts('y', P, cat='Binary')

# BIẾN MỚI: Biến cộng hưởng cho P8 và P13
z8_13 = LpVariable('z_synergy_8_13', cat='Binary')

# 4. Hàm mục tiêu: Tối đa hóa tổng lợi ích + Lợi ích cộng hưởng
m += lpSum(B[i] * y[i] for i in P) + B_extra * z8_13

# 5. Các ràng buộc (Constraints)
m += lpSum(C[i] * y[i] for i in P) <= 80000        # (C1) Ngân sách tổng 5 năm
m += lpSum(C1[i] * y[i] for i in P) <= 40000       # (C2) Ngân sách năm 1-2
m += y[1] + y[2] <= 1                              # (C3) Loại trừ Trung tâm dữ liệu
m += y[8] <= y[12]                                 # (C4) AI quốc gia cần đào tạo kỹ sư
m += y[13] <= y[12]                                # (C5) Khu bán dẫn cần đào tạo kỹ sư
m += y[4] + y[5] >= 1                              # (C6) Ít nhất một dự án Chính phủ số
m += y[14] >= 1                                    # (C6) An ninh mạng quốc gia là bắt buộc
m += lpSum(y[i] for i in P) >= 7                   # (C7) Số lượng dự án tối thiểu
m += lpSum(y[i] for i in P) <= 11                  # (C7) Số lượng dự án tối đa

# RÀNG BUỘC MỚI: Logic cho hiệu ứng cộng hưởng (Cổng AND tuyến tính)
m += z8_13 <= y[8], "Synergy_Link_P8"
m += z8_13 <= y[13], "Synergy_Link_P13"
m += z8_13 >= y[8] + y[13] - 1, "Synergy_Activation"

# 6. Giải bài toán
m.solve(PULP_CBC_CMD(msg=False))

# 7. Tính toán kết quả
selected_projects = [f"P{i}" for i in P if y[i].value() > 0.5]
synergy_active = "CÓ" if z8_13.value() > 0.5 else "KHÔNG"
total_cost = sum(C[i] for i in P if y[i].value() > 0.5)
total_benefit = value(m.objective)
marginal_npv = total_benefit / total_cost if total_cost > 0 else 0

# 8. Lưu trữ và báo cáo
output_dir = os.path.join("..", "outputs")
os.makedirs(output_dir, exist_ok=True)
output_file_path = os.path.join(output_dir, "ket_qua_cong_huong_bai5.txt")

report_content = (
    f"--- KẾT QUẢ TỐI ƯU CÓ HIỆU ỨNG CỘNG HƯỞNG (5.5.c) ---\n"
    f"1. Các dự án được chọn: {', '.join(selected_projects)}\n"
    f"2. Kích hoạt cộng hưởng (P8 & P13): {synergy_active}\n"
    f"3. Tổng chi phí đầu tư: {total_cost:,.0f} tỷ VND\n"
    f"4. Tổng lợi ích (Z* bao gồm cộng hưởng): {total_benefit:,.0f} tỷ VND\n"
    f"5. NPV biên (Z*/chi phí): {marginal_npv:.4f}\n"
)

print(report_content)

with open(output_file_path, "w", encoding="utf-8") as f:
    f.write(report_content)

print(f"[THÔNG BÁO] Kết quả cộng hưởng đã lưu tại: {os.path.abspath(output_file_path)}")
```
